```python
#Clean Table2
#Identify values of the headers of df2
header2 = list(df2.columns.values)
#Assign variable names to 2nd row 
for i in range(4):
  df2.iloc[0,i] = header2[i]
#create a new df only including "International migrant stock at mid-year (both sexes)"
#only column 2-9 were taken. column 1 does not provide any useful information
df2b = df2.iloc[:,1:10]
#The first row will replace the column headers & the first row will be dropped
df2b.columns = df2b.iloc[0]
df2b = df2b[1:]
# melt the value of years in the column header to reflect it as a value instead of variable
id_var_list = ['Major area, region, country or area of destination',
              'Notes',
              'Country code']
tidy_df2b = df2b.melt(id_vars = id_var_list, var_name = 'Year', value_name = 'Total population of both sexes at mid-year (thousands)')
#Create seperate column header "Sex" to assign value in column
tidy_df2b.insert(5,"Sex","Both")
#rename header title from 'International migrant stock at mid-year (both sexes)' to 'International migrant stock at mid-year'to 
tidy_df2b = tidy_df2b.rename(columns={'Total population of both sexes at mid-year (thousands)': 'Total population at mid-year (thousands)'})
#round to whole integer 
m=(tidy_df2b.dtypes=='float')
tidy_df2b.loc[:,m]=tidy_df2b.loc[:,m].astype(int)

#repeat for 'Total male population at mid-year (thousands)' subset
df2m = df2.iloc[:,1:16]
df2m.drop(df2m.iloc[:,3:9],axis =1, inplace=True)
df2m.columns = df2m.iloc[0]
df2m = df2m[1:]
id_var_list = ['Major area, region, country or area of destination',
              'Notes',
              'Country code']
tidy_df2m = df2m.melt(id_vars = id_var_list, var_name = 'Year', value_name = 'Total male population at mid-year (thousands)')
tidy_df2m.insert(5,"Sex","Male")
tidy_df2m = tidy_df2m.rename(columns={'Total male population at mid-year (thousands)': 'Total population at mid-year (thousands)'})
m=(tidy_df2m.dtypes=='float')
tidy_df2m.loc[:,m]=tidy_df2m.loc[:,m].astype(int)

#repeat for 'Total female population at mid-year (thousands)' subset
df2f = df2.iloc[:,1:22]
df2f.drop(df2f.iloc[:,3:15],axis =1, inplace=True)
df2f.columns = df2f.iloc[0]
df2f = df2f[1:]
id_var_list = ['Major area, region, country or area of destination',
              'Notes',
              'Country code']
tidy_df2f = df2f.melt(id_vars = id_var_list, var_name = 'Year', value_name = 'Total female population at mid-year (thousands)')
tidy_df2f.insert(5,"Sex","Female")
tidy_df2f = tidy_df2f.rename(columns={'Total female population at mid-year (thousands)': 'Total population at mid-year (thousands)'})

#consolidate three sub-tables
tidy_df2 =  pd.concat([tidy_df2f,tidy_df2m,tidy_df2b], ignore_index=True)
#remove decimals for "country code"
tidy_df2['Country code'] = tidy_df2['Country code'].astype(float).astype(int)
tidy_df2.tail()
```
