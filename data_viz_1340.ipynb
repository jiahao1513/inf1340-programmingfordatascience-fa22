```python
def table_transform(drop_cols, table):
    table.columns = table.columns.to_flat_index()
    table = table.drop(drop_cols,axis = 'columns')
    table.rename(columns = {('Country code', 'Unnamed: 3_level_1'):'Country code'},inplace = True)
    table.rename(columns = {('Major area, region, country or area of destination', 'Unnamed: 1_level_1'):'Major area'},inplace = True)
    table.set_index('Major area',inplace = True)
    return table
```

```python
drop_cols = [('Notes', 'Unnamed: 2_level_1'),('Type of data (a)', 'Unnamed: 4_level_1'),('Sort\norder', 'Unnamed: 0_level_1')]
```

```python
table1_new = table_transform(drop_cols,table1)
table3_new = table_transform(drop_cols,table3)
table4_new = table_transform(drop_cols,table4)
table5_new = table_transform(drop_cols,table5)
table6_new = table_transform(drop_cols,table6)
```

```python
drop_cols_table2 = [('Notes', 'Unnamed: 2_level_1'),('Sort\norder', 'Unnamed: 0_level_1')]
```

```python
table2_new = table_transform(drop_cols_table2,table2)
```
